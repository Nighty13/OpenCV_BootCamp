```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
hor = cv2.imread('DATA/PT10.jpg')
hor = cv2.cvtColor(hor,cv2.COLOR_BGR2RGB) 
plt.imshow(hor)
```

```python
#opencv : BGR order
#Blue Chan Hist
hist_values=cv2.calcHist([hor],channels=[0],mask=None,histSize=[256],ranges=[0,256])
```

```python
hist_values.shape
```

```python
plt.plot(hist_values)
```

```python
img = hor
```

```python
color = ('b','g','r')

for i,col in enumerate(color):
    histr=cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
    plt.xlim([0,256])
plt.title('IMG B G R HISTOGRAM')    
```

```python
gz = cv2.imread('DATA/PT07.jpg',0)
def display(img,cmap=None):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
gz = cv2.bilateralFilter(gz,5,35,35)
display(gz,cmap='gray')
```

```python
gz.shape
```

```python
hist_values = cv2.calcHist([gz],channels=[0],mask=None,histSize=[256],ranges=[0,256])
```

```python
plt.plot(hist_values)
hist_values.shape
```

```python
eq=cv2.equalizeHist(gz)
display(eq,cmap='gray')
```

```python
hist_values = cv2.calcHist([eq],channels=[0],mask=None,histSize=[256],ranges=[0,256])
plt.plot(hist_values)
```
