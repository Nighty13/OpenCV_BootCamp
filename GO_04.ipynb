```python
import cv2
from math import sin
import random
import numpy as np
```

```python
img = np.zeros((1000,1600,3),np.uint8)
B=G=R=rsize=0
time=0
des=der=1
it=0
stop=False
```

```python
def draw_circle(event,x,y,flags,param):
    global stop
    global img
    global B,G,R,rsize
    if event == cv2.EVENT_LBUTTONUP:
        stop=True
    if event == cv2.EVENT_LBUTTONDOWN:
        stop=False
    if(not stop):
        R=random.randrange(255)
        B=random.randrange(255)
        G=random.randrange(255)
        rsize=random.randrange(100)
    cv2.circle(img,(x,y),rsize,(B-180,G,R),-1)  
        
    
            

cv2.namedWindow(winname='my_drawing')
cv2.setMouseCallback('my_drawing',draw_circle)

while True : 
    #cv2.imshow('my_drawing',img)
    time=time+1
    if(it==2):
        if(der==10):
            der=1
        else:
            der=10
        it=0
    if(abs(time)>100):
        it=it+1
        des=des*-1
        time=0
    print(time)
    btx=random.randrange(-2,2)
    bty=random.randrange(-2,2)
    pts1 = np.float32([[5,sin(time)*des*der],[1600,0],[0,1000],[1600,1000]])
    pts2 = np.float32([[5,sin(time-1)*des],[1600,0],[0,1000],[1600,1000]])
    M = cv2.getPerspectiveTransform(pts1,pts2)
    img=cv2.warpPerspective(img,M,(1600,1000))
    grey=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    cv2.imshow('my_drawing',grey)
    if cv2.waitKey(20) & 0xFF == 27 : 
        break
cv2.destroyAllWindows()
```
