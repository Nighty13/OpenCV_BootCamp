```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img,cmap='gray'):
    fig=plt.figure(figsize=(12,10))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
cover=cv2.imread('DATA/PT16.jpg')
```

```python
display(cover)
```

```python
coverR=cv2.imread('DATA/PT18.jpg',0)
display(coverR)
```

```python
orb = cv2.ORB_create()
```

```python
kp1,des1=orb.detectAndCompute(cover,None)
kp2,des2=orb.detectAndCompute(coverR,None)
```

```python
bf = cv2.BFMatcher(cv2.NORM_HAMMING,crossCheck=True)
```

```python
matches = bf.match(des1,des2)
```

```python
single_match=matches[0]
```

```python
single_match.distance
len(matches)
```

```python
matches = sorted(matches,key=lambda x:x.distance)
```

```python
cover_matches = cv2.drawMatches(cover,kp1,coverR,kp2,matches[:100],None,flags=2)
```

```python
display(cover_matches)
```
