```python
img=cv2.imread('DATA/PT05.png',0)
```

```python
plt.imshow(img,cmap='gray')
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(10,10))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
show_pic(img)
img = cv2.bilateralFilter(img,3,9,9)
show_pic(img)
```

```python
ret,th1 = cv2.threshold(img,182,255,cv2.THRESH_BINARY)
show_pic(th1)
```

```python
th2=cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,25,5)
show_pic(th2)
th2=cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,8)
show_pic(th2)
```

```python
blended=cv2.addWeighted(src1=th1,alpha=0.6,src2=th2,beta=0.4,gamma=0)
```

```python
show_pic(blended)
```
