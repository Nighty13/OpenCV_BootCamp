```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('DATA/PT10.jpg')
imgr = cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
plt.imshow(imgr)
```

```python
edges = cv2.Canny(image=imgr,threshold1=100,threshold2=180)
plt.imshow(edges,cmap='gray')
```

```python
med_val = np.median(imgr)
med_val
```

```python
lower = int(max(0,0.7*med_val))
upper = int(min(0,0.7*med_val))
```

```python
edges = cv2.Canny(image=imgr,threshold1=lower,threshold2=upper)
plt.imshow(edges,cmap='gray')
```

```python
blurred_img = cv2.blur(imgr,ksize=(7,7))
edges = cv2.Canny(image=blurred_img,threshold1=lower,threshold2=upper)
plt.imshow(edges)
```

```python
blurred_img=cv2.GaussianBlur(imgr,(9,9),10)
edges = cv2.Canny(image=blurred_img,threshold1=lower,threshold2=upper)
plt.imshow(edges)
```

```python
blurred_img=cv2.medianBlur(imgr,9)
edges = cv2.Canny(image=blurred_img,threshold1=lower,threshold2=upper)
plt.imshow(edges)
```
