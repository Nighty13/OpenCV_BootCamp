```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess=  cv2.imread('DATA/PT11.jpg')
flat_chess=cv2.cvtColor(flat_chess,cv2.COLOR_RGB2BGR)
```

```python
plt.imshow(flat_chess)
```

```python
gray_flat_chess=cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_flat_chess,cmap='gray')
```

```python
real_chess=  cv2.imread('DATA/PT12.jpg')
real_chess=cv2.cvtColor(real_chess,cv2.COLOR_RGB2BGR)
plt.imshow(real_chess)
```

```python
gray_real_chess=cv2.cvtColor(real_chess,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_real_chess,cmap='gray')
```

```python
gray = np.float32(gray_flat_chess)
```

```python
dst = cv2.cornerHarris(src=gray,blockSize=2,ksize=5,k=0.04)
```

```python
dst = cv2.dilate(dst,None)
plt.imshow(dst,cmap='gray')
```

```python
flat_chess[dst>0.01*dst.max()]=[255,0,0]
```

```python
plt.imshow(flat_chess)
```

```python
gray = np.float32(gray_real_chess)
dst = cv2.cornerHarris(src=gray,blockSize=2,ksize=9,k=0.04)
plt.imshow(dst,cmap='gray')
```

```python
dst=cv2.dilate(dst,None)
plt.imshow(dst,cmap='gray')
```

```python
real_chess[dst>0.2*dst.max()]=[255,0,0]
plt.imshow(real_chess)
```

```python
real_chess=  cv2.imread('DATA/PT12.jpg')
real_chess=cv2.cvtColor(real_chess,cv2.COLOR_RGB2BGR)
flat_chess=  cv2.imread('DATA/PT11.jpg')
flat_chess=cv2.cvtColor(flat_chess,cv2.COLOR_RGB2BGR)
gray_flat_chess=cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
gray_real_chess=cv2.cvtColor(real_chess,cv2.COLOR_BGR2GRAY)
```

```python
corners = cv2.goodFeaturesToTrack(gray_flat_chess,128,0.01,10)
corners = np.int0(corners)
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(flat_chess,(x,y),3,(255,0,0),-1)
```

```python
plt.imshow(flat_chess)
```

```python
corners = cv2.goodFeaturesToTrack(gray_real_chess,120,0.01,10)
```

```python
corners = np.int0(corners)
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(real_chess,(x,y),5,(255,0,0),-1)
```

```python
plt.imshow(real_chess)
```
